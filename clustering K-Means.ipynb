```python
df.fillna(0,inplace=True)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["group"]=le.fit_transform(df["group"])
```

```python
# Encodage des variables categoriques

categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(col)
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df_scaled = scaler.fit_transform(df)
```

```python
print(df_scaled)
```

```python
from sklearn.decomposition import PCA

# Réduction de la dimension à 2D
pca = PCA(n_components=2)
df_reduced = pca.fit_transform(df_scaled)
```

```python
print(df_reduced)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

def optimise_kmeans(data, max_k) : 
    inertais = []
    for k in range(1,max_k+1) : 
        kmeans = KMeans(n_clusters=k) 
        kmeans.fit(data)
        inertais.append(kmeans.inertia_)
    return inertais 
```

```python
max_k = 10
inertais =  optimise_kmeans(df_reduced,max_k=max_k)
```

```python
fig = plt.subplots(figsize=(10,5))    
plt.plot(range(1,max_k+1), inertais, marker='o')
plt.xlabel('Nombre de clusters')
plt.ylabel('Inertais')
plt.grid(True)
plt.show()
```

```python

from sklearn.metrics import silhouette_score

# Calcul du Silhouette Score pour différents k
silhouette_scores = []
for k in range(2, max_k+1):  # Silhouette Score nécessite au moins 2 clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_reduced)
    score = silhouette_score(df_reduced, kmeans.labels_)
    silhouette_scores.append(score)

# Tracé du Silhouette Score
plt.figure(figsize=(10, 5))
plt.plot(range(2, max_k+1), silhouette_scores, marker='o')
plt.title('Silhouette Score')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
# Choix de k basé sur les scores
optimal_k = silhouette_scores.index(max(silhouette_scores)) + 2
print(f"Le nombre optimal de clusters selon le Silhouette Score est : {optimal_k}")
```

```python

kmeans = KMeans(n_clusters=optimal_k, random_state=42)  
kmeans.fit(df_reduced)

df['Cluster'] = kmeans.labels_
```

```python
# Analyse des clusters
df.groupby('Cluster').mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Cluster', data=df)
plt.title('Distribution des Clusters')
plt.show()
```
